```python
from nltk.tokenize import sent_tokenize
from konlpy.tag import Hannanum
from konlpy.tag import Kkma
from konlpy.tag import Okt

hannanum = Hannanum()
kkma = Kkma()
okt = Okt()
```

```python
text = "올시즌 37골로 시즌 골랭킹에서도 단연 으뜸인 호나우두는 이미 1992년 EPL 출범 이후 한 시즌 최다골 기록을 작성한 것은 물론 .하지만 빅4 외의 클럽들에는 유럽클럽 무대에 출전할 수 있는 절호의 기회로 인식되어 초반 라운드부터 베스트11을 총출동시킨다 ."
```
